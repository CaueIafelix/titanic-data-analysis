# **Persistencia do modelo**

```python
#importando o arquivo de teste
import pandas as pd
teste = pd.read_csv('titanic_test.csv')
```

```python
#Fazendo load do modelo
from joblib import load
reg = load('regrassao.joblib')
```

```python
#exibindo o intercept
reg.intercept_
```

```python
#exibindo o coef
reg.coef_
```

```python
#fazendo a previsão
teste = teste.drop('Unnamed: 0', axis=1)
X_teste = teste.drop('MedHouseVal', axis=1)

reg.predict(X_teste)
```

```python
#exportando esse arquivo
previsão = reg.predict(X_teste)

X_teste['previsão'] = previsão

X_teste.to_excel('previsão.xlsx')
```
